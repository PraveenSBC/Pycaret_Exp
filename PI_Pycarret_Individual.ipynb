#Mr.Beast

```python
# Load Dataset
df_final = pd.read_pickle('df_final_FE_@MrBeast.pkl')
df_final = df_final.sample(frac=1, random_state=42)


#Splitting Dataset for Training & Testing
data = df_final.sample(frac=0.8, random_state=42)
data_unseen = df_final.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))


# Get the list of numerical columns
numerical_columns = df_final.select_dtypes(include=['number']).columns.tolist()

# Get the list of categorical columns
categorical_columns = df_final.select_dtypes(exclude=['number']).columns.tolist()

numericals   = [ 'Previous_Video_Views', 'Previous_Video_Like','Previous_Video_Comment']
categoricals = []
ignore = categorical_columns
```

```python
# experience1 = no preprocessing
exp1 = RegressionExperiment()
exp1.setup(data, ignore_features=ignore,  target='#Views', session_id=555, train_size=0.7, preprocess=False, numeric_features=numericals,categorical_features=None)

top3_exp1 = exp1.compare_models(n_select=3)
```

```python
# Best Model
# Creating the Model
gradient_boost = exp1.create_model('gbr')
# Hyper parameter tuning
tune_gradient_boost = exp1.tune_model(gradient_boost)

# Predictions
# Non-Tuned
unseen_predictions = exp1.predict_model(gradient_boost, data=data_unseen)
# Tuned
unseen_predictions_tuned = exp1.predict_model(tune_gradient_boost, data=data_unseen)
```
